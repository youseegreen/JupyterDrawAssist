```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import Debug

debug = Debug.debug(True, wholeSize=(20,7), subNum=7)

for i in ['./imgs/sample1.png', './imgs/sample2.png']:
    # for
    debug.initialize()
    
    # 
    gray = cv2.imread(i, 0)
    color = cv2.imread(i, 1)
    debug.draw(gray, 'gray_img')
    debug.draw(color, 'color_img', 'brg')

    # 
    blur = cv2.GaussianBlur(gray,(19,19),0)
    debug.draw(blur, 'blur_img')

    # 
    binary = cv2.adaptiveThreshold(blur,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11,2)
    debug.draw(binary, 'binary_img')

    # 
    # 
    hist_img = cv2.calcHist([blur], [0], None, [256], [0, 256])
    debug.plot(hist_img, 'hist') 
```
